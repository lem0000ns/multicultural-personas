```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tools import utils

```

```python
# visualize bar graphs of ablation net accuracies
eng_y_easy = []
eng_y_hard = []
ling_y_easy = []
ling_y_hard = []
eng_x = []
ling_x = []
with open("../results/eng/country_distribution.json", "r") as f:
    data = json.load(f)["net_persona"]
    eng_baseline_easy = data["easy"]["Net Accuracy"]
    eng_baseline_hard = data["hard"]["Net Accuracy"]
with open("../results/ling/country_distribution.json", "r") as f:
    data = json.load(f)["net_persona"]
    ling_baseline_easy = data["easy"]["Net Accuracy"]
    ling_baseline_hard = data["hard"]["Net Accuracy"]
    
for mode in utils.modes_list:
    # for each mode, visualize easy and hard results
    if "eng" in mode and mode != "eng":
        eng_x.append(mode[7:])
        with open(f"../results/{mode}/country_distribution.json", "r") as f:
            data = json.load(f)["net_persona"]
        eng_y_easy.append(eng_baseline_easy - data["easy"]["Net Accuracy"])
        eng_y_hard.append(eng_baseline_hard - data["hard"]["Net Accuracy"])
    elif "ling" in mode and mode != "ling":
        ling_x.append(mode[8:])
        with open(f"../results/{mode}/country_distribution.json", "r") as f:
            data = json.load(f)["net_persona"]
        ling_y_easy.append(ling_baseline_easy - data["easy"]["Net Accuracy"])
        ling_y_hard.append(ling_baseline_hard - data["hard"]["Net Accuracy"])
```

```python
# English easy
ax = sns.barplot(x=eng_x, y=eng_y_easy, color="blue")
ax.set_xticklabels(eng_x, rotation=45)
ax.set_title("Impact on English Net Accuracy (Easy)")
ax.set_xlabel("Mode")
for i, v in enumerate(eng_y_easy):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom')
ax.set_ylabel("Net Accuracy")
```

```python
# English Hard Mode
ax = sns.barplot(x=eng_x, y=eng_y_hard, color="red")
ax.set_xticklabels(eng_x, rotation=45)
ax.set_title("Impact on English Net Accuracy (Hard)")
ax.set_xlabel("Mode")
for i, v in enumerate(eng_y_hard):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom')
ax.set_ylabel("Net Accuracy")
plt.show()
```

```python
# Ling Easy Mode
ax = sns.barplot(x=ling_x, y=ling_y_easy, color="green")
ax.set_xticklabels(ling_x, rotation=45)
ax.set_title("Impact on Ling Net Accuracy (Easy)")
ax.set_xlabel("Mode")
for i, v in enumerate(ling_y_easy):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom')
ax.set_ylabel("Net Accuracy")
plt.show()
```

```python
# Ling Hard Mode
ax = sns.barplot(x=ling_x, y=ling_y_hard, color="orange")
ax.set_xticklabels(ling_x, rotation=45)
ax.set_title("Impact on Ling Net Accuracy (Hard)")
ax.set_xlabel("Mode")
for i, v in enumerate(ling_y_hard):
    ax.text(i, v, f'{v:.2f}', ha='center', va='bottom')
ax.set_ylabel("Net Accuracy")
plt.show()
```
